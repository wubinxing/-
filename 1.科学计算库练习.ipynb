风格字符：-实现、--虚线、-.点划线、：点虚线、"空格"留空 空格

显示图例：
best:0  upper right:1  upper left:2   lower right:4   lower left:3   right:5  center left:6  center right:7  lower center:8  upper center:9   center:10

```python
import random

x = range(30)
y_shanghai = [random.uniform(10, 27) for i in x]

# nrows行， ncols列
fig, axes = plt.subplots(nrows = 2, ncols = 3, figsize = (20, 8), dpi = 100)

# plt.plot(x, y_shanghai)在最后一个坐标系展示

axes[0][1].plot(x, y_shanghai, linestyle = "--", 
                color = "r", label = "上海上午")
y_shanghai1 = [random.uniform(10, 27) for i in x]
axes[0][1].plot(x, y_shanghai1, linestyle = "-.", 
                color = "y", label = "上海下午")
# 显示图例
axes[0][1].legend(loc = 1) 
# 设置x轴刻度标签
x_ticks_label = ["10:{}".format(i) for i in x]
# print(x_ticks_label[::5])
# 构造y轴刻度
y_ticks = range(10, 30)
# AxesSubplot.set_xticks()
# AxesSubplot.set_xticklabels()
# plt.xticks(x[::2], x_ticks_label[::2])
axes[0][1].set_xticks(x[::5])
axes[0][1].set_xticklabels(x_ticks_label[::5], rotation=30)#rotation倾斜角度
axes[0][1].set_yticks(y_ticks[::5])
axes[0][1].set_yticklabels(y_ticks[::5])
# 画网格线
axes[0][1].grid(True, linestyle = "-.", color = "b",  alpha = 0.5)
# 设置
axes[0][1].set_xlabel("时间")
axes[0][1].set_ylabel("温度")
axes[0][1].set_title("上海时间温度变化图")
plt.show()
```
